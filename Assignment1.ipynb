```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('CO2.xlsx')
df = data.copy()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()  # To check any missing values in our data set
```

```python
df.info()  # To check the data types of the columns/features in the dataset
```

```python
# Sorting the data by population
df_pop = df.sort_values(by='Population', ascending=False)
```

```python
df_pop.head(10)  # Exploring the dataset
```

```python
# Bar chart for higest population in the world.

fig, ax = plt.subplots()

ax.bar(df_pop['Country'].head(10), df_pop["Population"].head(10), color='b')
ax.set_xticklabels(df_pop['Country'].head(10), rotation=90)
ax.set_facecolor("#c2f2e4")  # Changing the background color
ax.set_title("Top 10 Countries with highest population in the world ")
ax.set_ylabel("Values in 10^10  ")

plt.show()
```

```python
# Countries with highest C02 Emissions in the world.
df_c02 = df.sort_values(by='CO2Emissions', ascending=False)

fig, ax = plt.subplots()
ax.bar(df_c02['Country'].head(10), df_c02["CO2Emissions"].head(10), color='y')
ax.set_xticklabels(df_c02['Country'].head(10), rotation=90)
ax.set_facecolor("#f7f2bc")
ax.set_title("Top 10 Countries with highest C02 Emissions in the world ")
ax.set_ylabel("Values in tonnes 10^10  ")

plt.show()
```

```python
# A comparison between highest per capita and lowest per capita C02 emitters.
# Sorting data by percapita CO2 emission
df_per_c = df.sort_values(by='Percapita', ascending=False)

fig, ax = plt.subplots(1, 2, figsize=(17, 5))  # Setting the size of subplots


ax[0].bar(df_per_c['Country'].head(10), df_per_c["Percapita"].head(10), color='r')
ax[0].set_xticklabels(df_per_c['Country'].head(10), rotation=90)
ax[0].set_title("Top 10 Countries with highest per capita CO2 emissions ")
ax[0].set_ylabel("Values of per capita CO2 emission ")

ax[1].bar(df_per_c['Country'].tail(10), df_per_c["Percapita"].tail(10), color='g')
ax[1].set_xticklabels(df_per_c['Country'].tail(10), rotation=90)
ax[1].set_title("Top 10 Countries with lowest per capita CO2 emissions ")
ax[1].set_ylabel("Values of per capita CO2 emission ")

plt.show()
```

```python
# A boxplot to show the distribution of C02 and population distribution. There is a visible outlier in C02.

fig, ax = plt.subplots()
ax.boxplot([df["CO2Emissions"], df["Population"]])
ax.set_xticklabels(["Carbon dioxide emissions", "Country Population"])
ax.set_ylabel("Values in 10^10 ")
ax.set_title("Box-Plot for CO2 and Population", bbox={'facecolor': '0.8', 'pad': 5})
ax.set_facecolor("yellow")
plt.show()
```

```python
# Pie-chart to display the top 5 highest CO2 emitters

labels = df_c02["Country"].head()
sizes = df_c02['CO2Emissions'].head()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Top 5 highest CO2 emitters in the world", bbox={'facecolor': '0.8', 'pad': 5})
plt.show()
```

```python
labels = df_pop["Country"].head()
sizes = df_pop['Population'].head()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Top 5 most populous countries in the world", bbox={'facecolor': '0.8', 'pad': 5})
plt.show()
```

```python
# An interesing figure for countries having more life expectancy and their CO2 emissions

df_life = df.sort_values(by='LifeExpectancy', ascending=False)

fig, ax = plt.subplots(1, 2, figsize=(17, 5))

ax[0].bar(df_life['Country'].head(10), df_life["LifeExpectancy"].head(10), color='g')
ax[0].set_xticklabels(df_life['Country'].head(10), rotation=90)
ax[0].set_title("Top 10 Countries with highest Life Expectancy rate ")
ax[0].set_ylabel("Age ")

ax[1].bar(df_life['Country'].tail(10), df_life["LifeExpectancy"].tail(10), color='r')
ax[1].set_xticklabels(df_life['Country'].tail(10), rotation=90)
ax[1].set_title("Top 10 Countries with lowest Life Expectancy rate ")
ax[1].set_ylabel("Age ")
plt.show()
```
